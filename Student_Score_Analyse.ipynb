```python
import pandas as pd
df=pd.read_csv("marksheet (1).csv")
df.head()
```

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
df.head()
```

```python
df["Total Score"] = df[['Science','English','Maths','History']].sum(axis=1)
df["Average"]=df[['Science','English','Maths','History']].mean(axis=1)
df.head()
```

```python
def get_grade(avg):
    if avg>=90:
      return 'A+'
    elif avg>=75:
      return 'A'
    elif avg>=60:
      return 'B'
    elif avg>=50:
      return 'C'
    elif avg>=40:
      return 'D'
    else:
      return 'F'
df["Grade"]=df["Average"].apply(get_grade)
df.head()
```

```python
top_all_subjects=df[(df["Science"]>80) & (df["Maths"]>80) & (df["History"]>80) & (df["English"]>80)]
top_all_subjects[["Name","Science","English","History","Maths"]]
```
